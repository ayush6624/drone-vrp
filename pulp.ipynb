```python
from pulp import *
prob  = LpProblem("Diet", LpMinimize)
```

```python
x1 = LpVariable("Steak", 0, None, LpInteger)
x2= LpVariable("Peanut", 0 ,None, LpInteger)
```

```python
prob += 2 * x1 + 3 * x2, "Total Cost"
```

```python
prob += x1 + 2*x2 >= 4, "Minimum Protein Intake"
```

```python
prob.solve()
```

```python
print("Status: ", LpStatus[prob.status])
```

```python
value(prob.objective) 
```

# NetworkX

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
H = nx.Graph()
```

```python
for i in range(1, 6):
    H.add_node(i)
H.add_edge(1,2)
H.add_edge(1,3)
H.add_edge(2,3)
H.add_edge(4,6)
H.add_edge(5,4)
H.add_edge(2,6)

nx.draw(H, with_labels=1)
plt.show()
```

```python
H = nx.Graph()
cities = ['a', 'b', 'c', 'd', 'e']
for i in cities:
    H.add_node(i)
print(H.number_of_nodes())
# nx.draw(H, with_labels=1)
# plt.show()
# print(H.size())
```

```python

G = nx.Graph()
G.add_weighted_edges_from([("a", "b", 3.0), ("a", "c", 7.5), ("b", "c", 2.9)])

nx.draw(G, with_labels=True)
plt.axis("off")
plt.show()
```
